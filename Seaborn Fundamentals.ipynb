```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

seaborn is a python library for data visualization like as matplotlib


```python
survey = pd.read_csv("./data/young-people-survey-responses.csv" , index_col = 0)
survey
```

```python
plt.figure(figsize = (10 , 4))
plt.style.use("seaborn-colorblind")
sns.countplot(x = "Internet usage" , data = survey)
plt.show()
```

```python
plt.figure()
plt.style.use("seaborn-colorblind")
order = ['no time at all' , 'most of the day' , 'less than an hour a day' , 'few hours a day' ]
sns.catplot(x = "Internet usage" , data = survey , kind = 'count' , order = order)
plt.show()
```

```python
dataframe = pd.read_csv("./data/countries-of-the-world.csv" , index_col = 0)
dataframe
```

```python
dataframe['Region'].nunique()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style('darkgrid')
sns.scatterplot(x = 'GDP ($ per capita)' , y = 'Population' , data = dataframe , hue = 'Region')
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style('darkgrid')
sns.regplot(x = 'GDP ($ per capita)' , y = 'Population' , data = dataframe )
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style('darkgrid')
sns.lmplot(x = 'GDP ($ per capita)' , y = 'Population' , data = dataframe , hue = 'Region')
plt.show()
```

```python
plt.figure(figsize = (15 , 5))
sns.set_palette("RdBu")
sns.set_context("paper")
sns.set_style('darkgrid')
sns.stripplot(x = 'Region' , y = 'GDP ($ per capita)' , data = dataframe )
plt.title("Stripplot")
plt.xticks(rotation = 45)
plt.show()
```
