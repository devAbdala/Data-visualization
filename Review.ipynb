```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
sns.set_palette("PRGn")
sns.set_context("notebook")
```

```python
farmers = pd.read_csv("./data/US_farmers.csv" , index_col = 0)
farmers
```

```python
farmers[['state']].nunique()
```

```python
farmers[['state']].value_counts()
```

```python
plt.figure(figsize=(10 , 5))
sns.countplot(x = farmers['state'])
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(10 , 5))
sns.barplot(x = 'state' , y = 'Beans' , data = farmers)
plt.xticks(rotation = 90)
plt.show()
```

```python
alcohol = pd.read_csv("./data/student-alcohol-consumption.csv" , index_col = 0)
alcohol
```

```python
plt.figure(figsize=(10 , 5))
sns.stripplot(x = 'sex' , y = 'absences' , data = alcohol)
plt.show()
```

```python
plt.figure(figsize=(10 , 5))
sns.countplot(x = 'location' ,  data = alcohol , hue = 'sex' )
plt.show()
```

```python
states = pd.read_csv("./data/census-state-populations.csv" , index_col=0)
states
```

```python
plt.figure(figsize=(20 , 5))
sns.lineplot(data = states)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(20 , 5))
sns.histplot(data = states , bins = 50)
plt.show()
```

```python
plt.figure(figsize=(10 , 10))
sns.heatmap(data = states , annot = True)
plt.show()
```

```python
plt.figure(figsize=(20 , 5))
sns.kdeplot(data = states , fill = True)
plt.show()
```
