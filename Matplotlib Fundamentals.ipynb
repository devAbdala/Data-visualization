```python
# Adding values to axes (x , y) --> 2 dimensions (2d)
# visuale 2 plots in same figure (subplots)
# customize some features for figure and axes (title , x-axis-label , y-axis-label)
# customizing line plots 
# 1- markers(Adding markers (o , v , .) )
# 2- linestyles ( -- , none) but default (-) for street line
# 3- choosing a color
fig , ax = plt.subplots()
fig.set_size_inches(15 , 3)
# matplotlib for figure styles (bmh , ggplot , seaborn-colorblind , grayscale , seaborn-white , dark-background)
plt.style.use('ggplot')
ax.plot(seattle_weather['DATE'] , seattle_weather['MLY-TAVG-NORMAL'] , marker = 'o' , linestyle = '--' , color = 'red' , label = 'seattle') # lineplot
ax.plot(austin_weather['DATE'] , austin_weather['MLY-TAVG-NORMAL'] , marker = '.' , linestyle = '--' , color = 'green' , label = 'austin') # lineplot
ax.set_xlabel('MONTHS')
ax.set_ylabel('MLY-TAVG-NORMAL')
ax.set_title('Weather')
plt.legend()
plt.show()
```

```python
climate_change = pd.read_csv('./datasets/climate_change.csv' , index_col = 0 , parse_dates = True)
climate_change
```

```python
climate_change.index
```

```python
fig , ax = plt.subplots()
fig.set_size_inches(15 , 5)
ax.plot(climate_change.index , climate_change['co2'])
plt.show()
```

```python
seventies = climate_change.loc['1970-01-01' : '1979-12-31']
fig , ax = plt.subplots()
fig.set_size_inches(15 , 5)
ax.plot(seventies.index , seventies['co2'])
plt.show()
```

```python
seventy_three = climate_change.loc['1973-01-01' : '1973-12-31']
fig , ax = plt.subplots()
fig.set_size_inches(15 , 5)
ax.plot(seventy_three.index , seventy_three['co2'])
plt.show()
```

```python
fig , ax = plt.subplots()
fig.set_size_inches(15 , 5)
ax.set_title("Time Series data")
ax.set_xlabel("Time Series")
ax.set_ylabel("co2(ppm)")
ax.tick_params('y' , colors = 'red')
ax.plot(climate_change.index , climate_change['co2'] , color = 'red')
ax2 = ax.twinx()
ax2.set_ylabel("Relative Temp (Celsuis)")
ax2.tick_params('y' , colors = 'blue')
ax2.plot(climate_change.index , climate_change['relative_temp'] , color = 'blue')
plt.show()
```

```python
# creating function for plotting tme series data
def time_series (axes , x , y , color , xlabel , ylabel) :
    axes.plot(x , y , color = color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.tick_params('y' , colors = color)
```

```python
fig , ax = plt.subplots()
fig.set_size_inches(15 , 5)
time_series(ax , climate_change.index , climate_change['co2'] , 'red' , 'Time series' , 'co2(ppm)')
ax2 = ax.twinx()
time_series(ax2 , climate_change.index , climate_change['relative_temp'] , 'blue' , 'Time series' , 'relative_Temp(celsuis)')
plt.show()
```

```python
rio = pd.read_csv("./datasets/Rio2016.csv" , index_col = 0)
rio
```

```python
fig , ax = plt.subplots()
fig.set_size_inches(55 , 5)
plt.style.use('bmh')
ax.set_title('Average height all sports')
for Sport in rio['Sport'] :
    sport_df = rio[rio['Sport'] == Sport]
    ax.bar(Sport , sport_df['Height'].mean() , yerr = sport_df['Height'].std())
plt.show()
fig.savefig("./figures/Automate1.png" , dpi = 300 )
```

```python
fig , ax = plt.subplots()
fig.set_size_inches(55 , 5)
plt.style.use('bmh')
ax.set_title('Average weight all sports')
for Sport in rio['Sport'] :
    sport_df = rio[rio['Sport'] == Sport]
    ax.bar(Sport , sport_df['Weight'].mean() , yerr = sport_df['Weight'].std())
plt.show()
fig.savefig("./figures/Automate2.png" , dpi = 300 )
```
