```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg = pd.read_csv("./data/mpg.csv" , index_col = 0)
mpg
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn")
sns.set_context("poster")
sns.scatterplot(x = 'weight' , y = 'horsepower' , data = mpg , hue = 'cylinders')
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn")
sns.set_context("poster")
sns.scatterplot(x = 'weight' , y = 'horsepower' , data = mpg , hue = 'model_year')
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn")
sns.set_context("poster")
sns.scatterplot(x = 'weight' , y = 'horsepower' , data = mpg , hue = 'origin')
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn")
sns.set_context("poster")
sns.regplot(x = 'weight' , y = 'horsepower' , data = mpg)
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("RdBu")
sns.set_context("talk")
sns.swarmplot(x = 'origin' , y = 'weight' , data = mpg)
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn")
sns.set_context("notebook")
sns.lmplot(x = 'weight' , y = 'horsepower' , data = mpg , hue = 'origin')
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn")
sns.set_context("notebook")
sns.kdeplot(x = mpg['weight'] , data = mpg , fill = True )
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn_r")
sns.set_context("notebook")
sns.kdeplot(x = mpg['acceleration'] , data = mpg , fill = True , hue = 'origin')
plt.show()
```

```python
pollution = pd.read_csv("./data/Pollution.csv" , index_col=0)
pollution
```

```python
pollution.index
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn_r")
sns.set_context("notebook")
sns.scatterplot(x = 'NO2' , y = 'SO2' , data = pollution)
plt.text(30 , 138 , 'look at this outlier' , fontdict= {"ha" : "left" , "size" : "large"})
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn_r")
sns.set_context("notebook")
sns.scatterplot(x = 'NO2' , y = 'SO2' , data = pollution ,  hue = 'city')
plt.annotate('Look at this outlier for Cincinnati', xy = (30 , 105) , xytext = (40 , 120) , arrowprops = {"facecolor" : "gray" , "width" : 4} , backgroundcolor = 'white')
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn_r")
sns.set_context("notebook")
sns.barplot(y = pollution.index , x = 'CO' , data = pollution )
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn_r")
sns.set_context("notebook")
sns.barplot(x = pollution.index , y = 'NO2' , data = pollution )
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn_r")
sns.set_context("notebook")
sns.lineplot(x = pollution.index , y = pollution['NO2'] , hue = pollution['year'])
plt.show()
```

```python
plt.figure(figsize = (20 , 5))
sns.set_style("darkgrid")
sns.set_palette("PRGn_r")
sns.set_context("notebook")
sns.lineplot(x = pollution['month'] , y = pollution['NO2'] , hue = pollution['year'] , data = pollution)
plt.show()
```
